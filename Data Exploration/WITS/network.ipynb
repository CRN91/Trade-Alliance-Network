```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
trade_data = pd.read_csv('./data/trade_data.csv')
trade_data.info()
```

```python
network = nx.from_pandas_edgelist(trade_data, 'reporter', 'partner', 'export')
#plot network
plt.figure(figsize=(10,10))
nx.draw(network, with_labels=True, node_size=1000, node_color='skyblue', edge_color='black', linewidths=1, font_size=15)
plt.show()
```
